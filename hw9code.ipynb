```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theta=np.arange(0,.5*np.pi,.1)
b=2
```

```python
a=(np.sqrt(1-((np.sin(theta))**2)/b**2))/np.cos(theta)
```

```python
ri=(1-a*b)/(1+a*b)
ti=2/(1+a*b)
```

```python
plt.plot(theta, ti, 'b', label = 'Et/Ei')
plt.plot(theta, ri, 'r', label='Er/Ei')
plt.grid()
plt.legend()
plt.xlabel('theta in radians')
```

```python
ref=(1-a*b)**2/(1+a*b)**2
tran=(4*a*b)/(1+a*b)**2
```

```python
plt.plot(theta, tran, 'b', label = 'T')
plt.plot(theta, ref, 'r', label='R')
plt.grid()
plt.legend()
plt.xlabel('theta in radians')
```
