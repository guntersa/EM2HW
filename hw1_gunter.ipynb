```python
import numpy as np                
import matplotlib.pyplot as plt   
plt.style.use('ggplot')  
%matplotlib inline
```

```python
k=9e9
```

```python
rSource = np.array([0,0])         
rObs = np.array([0.1,0])          
Q=.1e-6

r = rObs - rSource                

E = k*Q*r/(np.linalg.norm(r))**3  


fig = plt.figure(figsize=(5,5))


for n in range(1,21):
    theta = n*2*(np.pi/20)
    x=.1*np.cos(theta)
    y=.1*np.sin(theta)
    
    robs1=np.array([x,y])
    r1=robs1-rSource
    
    E1 = k*Q*r1/(np.linalg.norm(r1))**3  
    Escale1 = np.linalg.norm(r1)/np.linalg.norm(E1)/2      
    plt.plot(0,0,'ro')
    plt.arrow(robs1[0],robs1[1],Escale1*E1[0],Escale1*E1[1], color='green')  

plt.axis([-.2,.2,-.2,.2])
```

```python
qp=.1e-6
qn=-.1e-6

rsourcep=np.array([0, -.01])
rsourcen=np.array([0, .01])

fig = plt.figure(figsize=(5,5))

plt.plot(rsourcep[0], rsourcep[1], 'ro')
plt.plot(rsourcen[0], rsourcen[1], 'bo')


for n in range(1,41):
    theta=n*(2*np.pi/40)
    x=.1*np.cos(theta)
    y=.1*np.sin(theta)
    
    robs3=np.array([x,y])
    rpos=robs3-rsourcep
    rneg=robs3-rsourcen
    
    E3=(k*qp*rpos/(np.linalg.norm(rpos))**3) + (k*qn*rneg/(np.linalg.norm(rneg))**3)  
    Escale3 = np.linalg.norm(rpos)/np.linalg.norm(E3)/4      

    plt.arrow(robs3[0],robs3[1],Escale3*E3[0],Escale3*E3[1], color='green')  

plt.axis([-.2,.2,-.2,.2])
```
