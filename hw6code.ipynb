```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w=np.arange(.1,50,.1)
c=1
l=1
r=1
phi=np.arctan((r*w*c)-(r/(w*l)))
v=1
f=w/(2*np.pi)
I=(v*np.sqrt(((1/r)**2)+(w*c-(1/(w*l)))**2))*np.cos(w-phi)

plt.plot(f,I)
plt.grid()
```

The current reaches an extremum about every .5 Hz
